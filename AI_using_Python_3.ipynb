###**Tpyes of Data**
### **Structured Data**
This is highly organized data that fits neatly into fixed fields and formats, like a spreadsheet.
* Format: Rows and columns (Tabular).
* Storage: Relational Databases (SQL).
* Examples: Names, dates, credit card numbers, and stock prices.

### **Unstructured Data**
This is data that has no predefined format or organization. It is usually "heavy" and makes up about 80% of all enterprise data.
* Format: Random, various formats.
* Storage: Data Lakes or NoSQL databases.
* Examples: Emails, social media posts, videos, audio files, and satellite images.
* Searchability: Difficult to search without using AI or specialized tools.
### **Semi-Structured Data**
This is a "middle ground." It doesn’t fit into a rigid grid like a spreadsheet, but it contains tags or markers that separate data elements and provide some level of organization.
* Format: Hierarchical or graph-based.
* Storage: XML or JSON files.

* Searchability: Easier to process than unstructured data because of its internal labels.

###**High Quality Data**
High-quality data is information that is fit for its intended purpose in operations, decision-making, and planning. In data science, data is considered high quality when it correctly represents the real-world entities or events it is meant to describe.

Organizations typically measure data quality using six core "dimensions":
1. Accuracy: The degree to which data correctly reflects real-world facts or events (e.g., a customer's address matches their actual residence).
2. Completeness: Ensuring all required data is present and there are no missing values in critical fields, such as a patient's medical history.
3. Consistency: The uniformity of data across different systems. For example, a customer's phone number should be identical in both sales and shipping databases.
4. Timeliness: Whether the data is up-to-date and available when needed for making decisions (e.g., real-time stock prices for trading).
5. Validity: The degree to which data conforms to predefined formats, rules, or standards, such as a birthdate following a specific DD/MM/YYYY pattern.
6. Uniqueness: Ensuring each record is distinct and not duplicated within a dataset, which prevents inflated metrics or redundant interactions.

###**High Quality Data**
* **Relevance:** The data must directly relate to the specific problem you are trying to solve. Including "noise" or unrelated variables (like using a person's favorite color to predict their credit score) can confuse the model and decrease performance.
* **Accuracy:** The data must be correct and free from errors. If the "labels" or values are wrong (e.g., a cat labeled as a dog), the model will learn incorrect patterns, leading to flawed predictions.
* **Diversity:** The dataset should represent all possible real-world scenarios, including "edge cases." A lack of diversity leads to bias—for example, a facial recognition system that only sees one ethnicity will fail to work for others.
* **sufficiency:** You need enough data to capture the underlying patterns. If the data is too sparse, the model might "overfit" (memorize the few examples it has) instead of actually learning how to make predictions on new information.


###**Role of Data in ML Pipline**
* **Collection:** The foundation where raw data is gathered from various sources (databases, APIs, sensors, or web scraping). The goal is to ensure the data is relevant and diverse enough to cover all possible real-world scenarios the model might encounter.

* **Pre-processing:** The "cleaning" phase. Raw data is often messy, so this step involves handling missing values, removing duplicates, correcting errors, and formatting the data so it’s readable by the algorithm. It turns "garbage" into usable input.
* **Feature Engineering:** The creative process of selecting and transforming raw variables into more meaningful "features" that help the model learn better. This includes scaling numbers (making sure they are in the same range) or creating new data points (like extracting the "hour" from a timestamp).
* **Model Training:** Data is split into sets (usually training and validation). The algorithm "looks" at the training data to find patterns and correlations, adjusting its internal parameters to minimize errors based on the examples provided.

* **Evaluation:** The "test" phase. The model is shown unseen data to see how it performs in the real world. This step uses metrics like accuracy or precision to verify if the model actually learned the patterns or just "memorized" the training set.

In NumPy, indexing and slicing are used to access and extract specific elements or sub-sections of an array. Strides define how NumPy "steps" through the memory to find those elements.
###**Indexing (Accessing Single Elements)**
NumPy uses zero-based indexing. You can access elements using square brackets [].
* 1D Array: arr[i] (index i).
* 2D Array: arr[row, col] (using a comma to separate dimensions).
### **Slicing (Extracting Sub-sections)**
Slicing allows you to pull out a "slice" of an array using the syntax [start:stop:step].
* start: Inclusion index (default is 0).
* stop: Exclusion index (goes up to, but not including).
* step: The interval between elements.
###**Strides (The Engine Under the Hood)**
Strides is an internal NumPy attribute that tells the computer how many bytes it must skip in memory to move to the next element along each axis.
* It allows NumPy to create "views" of data without copying it.
* When you slice an array, NumPy doesn't create a new array; it just changes the strides to skip certain numbers, making it incredibly memory-efficient.
####**Summary of Syntax:**
* arr[index] : Access one item.
* arr[start:stop] : Access a range.
* arr[::step] : Access with a gap (e.g., ::-1 reverses the array).
* arr[row_start:row_stop, col_start:col_stop] : Slice a 2D matrix.

```python
arr = np.array([[10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]])

print(arr[0, 1])   # 20 (row 0, col 1)
print(arr[2, 2])   # 90 (row 2, col 2)

print(arr[-1, -2])  # 80 (last row, 2nd last column)
print(arr[-3, -3])  # 10 (first row, first column)

# General form: arr[start:stop:step]
print(arr[0:2, 1:3])   # rows 0-1, cols 1-2

print(arr[::-2, ::2])  # every 2nd row & 2nd column

slice_arr = arr[0, :2]  # view (not a copy)
slice_arr[0] = 999

print(arr[0])    # original changes

copy_arr = arr[1].copy()  # independent copy
copy_arr[0] = 555

print(arr[1])    # original unchanged

# Random generator with seed (reproducibility)
rng = np.random.default_rng(seed=42)
num = rng.integers(1,10, size=5)
print(num)
```

### **NumPy Random Numbers**
In NumPy, generating random numbers is essential for simulations and machine learning, but managing how they are generated is key to ensuring your results can be replicated by others.
### **The Modern Random Generator**
As seen in your previous code, the current best practice is to use np.random.default_rng(). This replaces the older np.random.rand functions.

###**Reproducibility (The "Seed")**
In data science, you often need to get the same random numbers every time you run your script (e.g., to debug a model or share results). This is called reproducibility.
* By setting a seed, you ensure the "random" sequence is the same every time.
* Common Seeds: 42, 0, or 1.
###**Performance Tips**
NumPy is fast, but you can make it even faster by following these rules:
* Avoid Python Loops: Never use a for loop to perform math on an array. Use Vectorization instead.

Slow: [x + 2 for x in my_list]

Fast: my_array + 2

* Use the Right dtype: Using float32 instead of float64 can cut your memory usage in half and speed up computations if high precision isn't required.
* Pre-allocate Memory: If you know the size of your final array, create it first using np.zeros() or np.empty() and then fill it, rather than using np.append() (which creates a full copy of the array every time).
* In-Place Operations: Use arr += 1 instead of arr = arr + 1 to modify the array in its existing memory location rather than creating a new one.

##**Pandas Indexing & Selection - loc & iloc**
In Pandas, loc and iloc are the primary ways to select data from a DataFrame. The main difference is whether you are using labels or positions.
###**1. loc (Label-based Selection)**
loc is used to select rows and columns by their names (labels).
* Syntax: df.loc[row_label, column_label]
* Slicing: Unlike standard Python, the end index in loc is inclusive.
* Use case: Best when you know the names of the rows or columns you want.

###**2. iloc (Integer-based Selection)**
iloc is used to select data by its integer position (index starting from 0).
* Syntax: df.iloc[row_position, column_position]
* Slicing: Follows standard Python rules; the end index is exclusive.
* Use case: Best when you don't care about labels and just want the "first 5 rows" or "last 2 columns."

image.png


```python
# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'City': ['New York', 'Paris', 'London', 'Berlin'],
    'Salary': [70000, 80000, 60000, 90000]
}

df = pd.DataFrame(data)

print('Original DataFrame:\n', df)

# 1. Label-based selection (.loc)
subset_loc = df.loc[1:3, ['Name', 'Salary']]
print('Subset with .loc:\n', subset_loc)

# 2. Integer-based selection (.iloc)
subset_iloc = df.iloc[0:2, 1:3]
print('Subset with .iloc:\n', subset_iloc)

# 3. Boolean filtering
high_salary = df[df['Salary'] > 75000]
print('High Salary Filter:\n', high_salary)

# 4. Safe assignment using .loc
df.loc[df['Age'] < 30, 'Salary'] = 65000
print('Low Age Salary Change:\n', df.loc[df['Age'] < 30, 'Salary'])
```

##**Handling Missing Data**
In data science, missing values are usually represented as NaN (Not a Number). Handling them is one of the most important steps in data cleaning because most Machine Learning models cannot process datasets containing missing data.
Pandas provides three primary ways to manage these gaps:
###**1. Detecting Missing Data**
Before fixing anything, you need to see where the holes are.
* df.isnull(): Returns a table of True/False values.
* df.isnull().sum(): The most common way to check; it gives you a count of missing values for every column.
###**2. Dropping Missing Data (dropna)**
If a row or column has too many missing values to be useful, you can simply remove it.
* df.dropna(): Removes any row that contains at least one missing value.
* df.dropna(axis=1): Removes any column that contains a missing value.
* df.dropna(subset=['Age']): Only drops rows where the 'Age' column is missing.
###**3. Filling Missing Data (fillna)**
Instead of deleting data (which can lead to losing valuable information), you can "impute" or fill in the gaps.
* Constant Value: df['Score'].fillna(0) (replaces NaN with 0).
* Statistical Mean/Median: df['Age'].fillna(df['Age'].mean()) (replaces NaN with the average age).
* Forward/Backward Fill:

df.ffill(): Carries the last valid observation forward to the next row (great for time-series data).
###**4. Replacing Specific Values**
* Sometimes data isn't missing but is recorded as "Unknown" or "?".

```python
import numpy as np
import pandas as pd

df = pd.read_csv(
    "retail_sales.csv",
    parse_dates=["Date"],
    dtype={
    "Category": "category",
    "Region" : "category"
    }
)

# convert Date to Column to datetime
df.head()
```

```python
# Detect missing values
print(df.isna().sum())
```

```python

# Fill missing Age with mean
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mode()[0])

#Fill missing value with Median
df['Sales'] = df['Sales'].fillna(df['Sales'].median())

#Fill catagorical value with mode
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
```

##**Sorting & Ranking Data in Pandas**
In Pandas, sorting and ranking allow you to organize your data to find top performers, identify trends, or prepare your dataset for visualization.
###**1. Sorting Data (sort_values)**
The sort_values() method is the standard way to reorder your DataFrame based on one or more columns.

average: Assigns the average rank to ties (default).

min: Assigns the lowest rank to the group (e.g., two people tied for 1st both get "1").

dense: Similar to min, but the next rank after a tie is always the next integer (1, 1, 2...).

##**Merging & Joining DataFrames in Pandas**
In Pandas, you often need to combine data from different sources. This is done using Merge, Join, or Concatenate.
###**1. Merging (pd.merge)**
Merging is the most flexible way to combine DataFrames based on common columns (similar to a SQL JOIN).
* Inner Join (Default): Keeps only the rows where the "key" exists in both tables.
* Left Join: Keeps all rows from the left table and adds matching data from the right.
* Right Join: Keeps all rows from the right table and adds matching data from the left.
* Outer Join: Keeps all rows from both tables, filling missing values with NaN.

###**2. Joining (df.join)**
Joining is a specialized version of merging that combines DataFrames based on their index (row labels) rather than columns. By default, it performs a Left Join.

###**3. Concatenating (pd.concat)**
Concatenation simply "stacks" DataFrames on top of each other (vertically) or side-by-side (horizontally). It doesn't look for matching keys; it just glues them together.
* Vertical (Default): pd.concat([df1, df2]) — Useful for adding new rows.
* Horizontal: pd.concat([df1, df2], axis=1) — Useful for adding new columns.

```python
import pandas as pd

# Sample DataFrames
customers = pd.DataFrame({
    'CustomerID': [1,2,3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

orders = pd.DataFrame({
    'OrderID': [101,102,103,104],
    'CustomerID': [1,2,2,3],
    'Amount': [-250, 150, 200, 300]
})

# Inner merge on CustomerID
df_inner = pd.merge(customers, orders, on='CustomerID', how='inner')
print("Inner Merge:\n", df_inner)

# Left merge
df_left = pd.merge(customers, orders, on='CustomerID', how='left')
print("\nLeft Merge:\n", df_left)

# Concatenate rows example
df_concat = pd.concat([customers, customers])
print("\nConcatenated Rows:\n", df_concat)
```

##**Applying Functions - apply, map & pipe in Pandas**
In Pandas, these three methods allow you to perform custom operations on your data. They differ in scope (whether they act on a single value, a column, or the whole table).
###**1. map() (Element-wise for Series)**
The map() method is used only on Series (single columns). It is primarily used for substituting each value in a column with another value based on a dictionary or a function.
* Best for: Categorical mapping (e.g., changing "M/F" to "Male/Female").
* Syntax: df['column'].map(arg)

###**2. apply() (Flexible for Series & DataFrames)**
apply() is the "Swiss Army Knife" of Pandas. It can run a function across a whole column or across rows.
* On a Series: Acts on every individual element.
* On a DataFrame: Can act on entire rows (axis=1) or entire columns (axis=0).
* Syntax: df.apply(function)


###**3. pipe() (Table-wise / Method Chaining)**
pipe() is used to apply a function to the entire DataFrame. It is best used for "clean code" and method chaining, allowing you to pass the results of one custom function directly into another.
* Best for: Complex, multi-step data cleaning pipelines.
* Syntax: df.pipe(function)

##**EDA Key Questions & Mindset**
Exploratory Data Analysis (EDA) is the "detective work" of data science. Your mindset should be one of skepticism and curiosity: don't trust the data until you've interrogated it.
##**The EDA Mindset**
* Don't assume data is clean: Always expect missing values, typos, and outliers.
* Look for the "Why": If you see a spike in sales, don't just note it—ask if it was a holiday, a promotion, or a data entry error.
* Visualize first: Statistics (like the mean) can be misleading. A graph often reveals patterns (like clusters or bimodal distributions) that numbers hide.
##**Key Questions to Ask Your Data**
###**1. The Structure: "What am I looking at?"**
* How many rows and columns are there? (df.shape)
* What are the data types? Are numerical values accidentally stored as strings? (df.info())
* What does each row represent (an invoice, a user, a sensor reading)?
###**2. The Quality: "Can I trust this data?"**
* How much data is missing? Is it missing randomly, or is there a pattern? (df.isnull().sum())
* Are there duplicate records?
* Are the values within a realistic range? (e.g., is there a "negative" price or a 200-year-old user?)
###**3. The Distribution: "What is the 'shape' of my variables?**
* "What is the average (mean) vs. the middle value (median)? If they are far apart, your data is skewed.
* Are there outliers? Are they legitimate extremes or errors?
* Is the data balanced? (e.g., in a "Churn" dataset, do 99% of people stay? This makes prediction harder.)
###**4. The Relationships: "How do variables interact?"**
* Does variable X increase when variable Y increases? (Correlation)
* How do categorical labels (like "Gender" or "City") affect numerical outcomes (like "Spending")?
* Are there "Hidden Groups" in the data?
###**5. The Goal: "Is this data useful for my target?"**
* Which features have the strongest relationship with the outcome I want to predict?
* Do I need to create new features (Feature Engineering) to make the data more powerful?
##**A Simple EDA Checklist**
* Preview: head(), tail(), sample()
* Verify: info(), describe(), nunique()
* Clean: Handle NaNs and duplicates.
* Visualize: Use Histograms (for distribution) and Scatter plots (for relationships).

In EDA, Descriptive Statistics summarize what your data looks like, while Data Distributions show how the values are spread out.
###**1. Descriptive Statistics (The "Numbers")**
These measures help you understand the central tendency and the spread of your data.

Measures of Center:
* Mean: The average. Highly sensitive to outliers (one billionaire in a room makes everyone look rich).
* Median: The middle value. Great for skewed data (like house prices or salaries).
* Mode: The most frequent value. Best for categorical data (like "City" or "Color").

Measures of Spread:
* Standard Deviation: How far, on average, the numbers are from the mean. High SD = high volatility.
* Range: The difference between the Max and Min.
* Quartiles (IQR): The values at 25%, 50% (median), and 75%. Used to create box plots and find outliers.

Pandas Shortcut: df.describe() gives you almost all of these in one table.

###**2. Data Distributions (The "Shape")**
* Understanding the "shape" of your data tells you which statistical tests or ML models you can use.
* Normal Distribution (Bell Curve): Data is symmetrical around the mean. Most natural phenomena (height, IQ) follow this.Skewed Distribution:

Right (Positive) Skew: A long tail to the right (e.g., Household income; most people earn low/mid, few earn millions).

Left (Negative) Skew: A long tail to the left (e.g., Age of retirement).
* Bimodal Distribution: The data has two "peaks." This usually means you are looking at two different groups mixed together (e.g., heights of men and women combined).

##**Outlier**
In data science, an outlier is a data point that differs significantly from other observations. These can be caused by measurement errors, data corruption, or genuine extreme variations.

There are two primary statistical methods for detection and removal:
###**1. The Interquartile Range (IQR) Method**
This is the best method for skewed data (data that doesn't follow a perfect bell curve). It uses the "middle 50%" of the data to define what is "normal."
* Step 1: Calculate IQR: \(IQR = Q3 (75th\ percentile) - Q1 (25th\ percentile)\)
* Step 2: Define Bounds:

Lower Bound = \(Q1 - (1.5 \times IQR)\)

Upper Bound = \(Q3 + (1.5 \times IQR)\)
* Detection: Any value below the Lower Bound or above the Upper Bound is an outlier.
###**2. The Z-Score Method**
This is best for Normally Distributed (Gaussian) data. It measures how many standard deviations a point is from the mean.
* Rule: Most data (99.7%) falls within 3 standard deviations.
* Detection: Any point with a Z-score > 3 or < -3 is typically considered an outlier.
*Formula: Z = (Value - Mean)/ Standard Deviation

Outlier Removal MethodsOnce detected, you have three main ways to handle them:
* Trimming (Dropping)
* Capping (Winsorization)
* Transformation

```python
# Load dataset
import pandas as pd

df = pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=["order_date"])

# IQR Method
Q1 = df["revenue"].quantile(0.25)
Q3 = df["revenue"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df["revenue"] < lower_bound) | (df["revenue"] > upper_bound)]

#IQR (Interquartile Range): Measures the middle 50% of data. Robust for skewed distributions (like revenue)

outliers_iqr.shape    #Verification
```

```python
outliers_iqr[["revenue", "product_category"]].head()    #Inspect
```

```python
df = df[(df["revenue"] >= lower_bound) & (df["revenue"] <= upper_bound)]
df.shape
```

```python
# Z-Score Method
from scipy.stats import zscore

df = pd.read_csv("ecommerce_sales_analytics_5000.csv", parse_dates=["order_date"])
df["revenue_z"] = zscore(df["revenue"])
outliers_z = df[df["revenue_z"].abs() > 3]

#Z-score shows: How many standard deviations a value is from the mean
# Rule: |Z| > 3 → outlier

outliers_z.shape    #Verification
```

```python
df = df[df["revenue_z"].abs() <= 3]
df.shape
```

In data science, Covariance and Correlation are used to measure the relationship between two variables. They tell you if, and how strongly, variables move together.
###**Covariance**
Covariance indicates the direction of the linear relationship between two variables.
* Positive Covariance: Both variables move in the same direction (if X goes up, Y tends to go up).
* Negative Covariance: Variables move in opposite directions (if X goes up, Y tends to go down).
* Zero Covariance: No linear relationship exists.
* The Problem: The value is hard to interpret because it depends on the scale (units) of the data. For example, covariance in "meters" will be different from "kilometers" for the same data.
###**Correlation (Pearson’s r)**
Correlation is the standardized version of covariance. It tells you both the direction and the strength of the relationship.
* Range: Always between -1.0 and +1.0.

+1.0: Perfect positive relationship.

-1.0: Perfect negative relationship.

0: No linear relationship.
* The Benefit: Because it is a standardized scale, you can compare the relationship between any two variables (e.g., height vs. weight) regardless of their units.

##**Creating an Exploratory Data Analysis Report**
An Exploratory Data Analysis (EDA) Report is a structured document that summarizes your initial findings from a dataset before you move on to complex modeling. It acts as a "quality check" and a roadmap for the rest of your project.
###**1. Standard Report Structure**
A professional report typically moves from a broad overview to specific insights:
* Executive Summary: A high-level view of the primary business question, the most important headline insights, and recommended next steps for stakeholders.
* Data Overview & Metadata: A "snapshot" of the data, including the number of rows/columns, source info, and a data dictionary defining what each variable means.
* Data Quality Report: A summary of technical issues found and how they were handled:

Missingness Map: Percentages of missing values and your imputation strategy.

Duplicate & Key Integrity: Confirmation that unique IDs are actually unique and duplicates have been removed.

Outlier Log: List of extreme values detected (via IQR or Z-score) and whether they were kept, capped, or removed.
* Univariate Analysis: Individual looks at key variables using histograms (for shape) and box plots (for spread).
* Bi/Multivariate Analysis: Examining how variables interact, typically featuring correlation heatmaps and scatter plots.
* Conclusion & Modeling Plan: Final takeaways, including which features look most promising and any new features you created (Feature Engineering).
###**2. Automated EDA Tools (The "Fast Way")**
If you need a report quickly, Python has libraries that generate comprehensive, interactive HTML reports with just one or two lines of code:
* YData-Profiling: Best for in-depth statistical summaries and data quality alerts.
* Sweetviz: Excellent for visually appealing reports and comparing two datasets (e.g., Train vs. Test).
* D-Tale: Provides a highly interactive web interface similar to an Excel sheet inside your browser.
* DataPrep: One of the fastest tools for generating reports on very large datasets.
###**3. Best Practices for Reporting**
* Lead with the Conclusion: Put your most important finding first so senior stakeholders don't have to hunt for it.
* Context is Key: Don't just show a graph; explain why it matters to the business goal.
* Avoid "Analysis Paralysis": Focus on the 3–5 most impactful variables rather than trying to explain every single column in the dataset.
* Tailor the Language: Use technical details for your peer data scientists but plain language for non-technical managers.

###**6-8 bullet EDA Summary**
An effective EDA Summary should capture the "vital signs" of your dataset and the specific actions taken to prepare it. Here is a 6-8 bullet template:
* Dataset Dimensions: State the final count of rows and columns after initial cleaning (e.g., "Analyzed 50,000 records across 15 features").
* Data Quality & Integrity: Summarize how missing values (NaNs) and duplicates were handled (e.g., "Imputed 5% missing 'Age' data using median values; removed 200 duplicate entries").
* Key Statistical Insights: Highlight the central tendency and spread of your target variable (e.g., "Average monthly sales are $12k, with a right-skewed distribution indicating a few high-value outliers").
* Outlier Treatment: Define the threshold used and the result (e.g., "Identified 12 extreme outliers using the 1.5x IQR rule; capped values to the 99th percentile to reduce model variance").
* Top Correlations: Identify the strongest relationships found (e.g., "Strong positive correlation (\(r=0.85\)) found between 'Marketing Spend' and 'User Acquisition'").
* Categorical Breakdowns: Note major imbalances in groups (e.g., "Dataset is imbalanced; 85% of users are in the 'Standard' tier, while only 15% are 'Premium'").
* Feature Engineering: List any new variables created (e.g., "Extracted 'Day of Week' from timestamps and created a 'Price-per-Unit' feature").
* Modeling Recommendation: Conclude with which features show the most predictive power for the next phase.
