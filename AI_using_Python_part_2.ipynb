# **Object-Oriented Programming (OOPs)**
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects," which can contain data (attributes) and code (methods).
Instead of writing a long list of instructions (procedural), you create blueprints called Classes to build individual Objects.

## **The 4 Pillars of OOP**
1. Inheritance: A new class can derive properties from an existing class (e.g., a Car inherits from Vehicle).
2. Encapsulation: Hiding the internal details of how an object works and protecting its data.
3. Polymorphism: Allowing different classes to be treated as instances of the same general class (e.g., both Dog and Cat have a speak() method).
4. Abstraction: Hiding complex logic behind simple interfaces (e.g., you press a "Start" button without needing to know how the engine combustion works).

## **Need of OOP**
### **Code Reusability :**
You can use "Inheritance" to avoid writing the same code for similar objects.

### **Scalability :**
It’s much easier to manage thousands of lines of code when it's organized into logical objects.

### **Maintainability**
If a bug exists in the "User" logic, you only look at the User class, not the whole program.

### **Data Security :**
Encapsulation prevents accidental changes to critical data from outside the class.

### **Real-world Modeling :**
It allows programmers to map code to real-life entities (like Users, Products, or Bank Accounts) easily.

## **Simple Comparison**
* Without OOP: You have separate variables like car1_color, car1_brand, car2_color, car2_brand.
* With OOP: You have a Car class, and you just create car1 = Car("Red", "Toyota").



## **In Python, OOP is built on these four fundamental building blocks:**
### 1. **Classes (The Blueprint)**

A Class is a logical template or "cookie cutter" used to create objects. It defines what data an object will hold and what it can do, but it doesn't contain actual data itself.

Syntax: class Dog:

### 2. **Objects (The Instance)**

An Object is a specific "copy" created from the class. While the class is the blueprint for a "Dog," the object is the actual dog named "Buddy."

Syntax: my_dog = Dog()

### 3. **Attributes (The Data)**

Attributes are variables that belong to an object. they represent the state or characteristics of the object.

* Example: A Car object might have attributes like color, brand, and max_speed.

* The __init__ Method: This is a special "constructor" used to set these attributes when an object is first created.
### 4. **Methods (The Behavior)**

Methods are functions defined inside a class. They represent the actions an object can perform.
* **The self Parameter:** Every method must include self as its first argument, which allows the method to access the attributes of the specific object that called it.


```python
class Car:
  def __init__(self,color):
    self.color = color

  def start(self):
    print(f'{self.color} car started')

# First Object
my_car = Car('red')
my_car.start()

# Second Object
my_brother_car = Car('blue')
my_brother_car.start()



class Car:
  def driver(self):
    print('The car is moving')

my_car = Car()
my_car.driver()


class Car:
  color = 'Red'
  def driver(self):
    print(f'{self.color} car is moving')

my_car = Car()
my_car.driver()


class Car:
  color = 'Red'
  def driver(self):
    print(f'{self.color} car is moving')

  def set_color(self,new_color):
    self.color = new_color

my_car_2 = Car()
my_car_2.set_color('Blue')
my_car_2.driver()
```

# **The __init__() Constructor Method**
The __init__ method is a special "magic" or "dunder" (double underscore) method in Python that acts as the constructor. It is automatically called every time you create a new object from a class.

## **Purpose**
Its primary job is to initialize (set the starting values) of an object's attributes. Without __init__, every object you create from a class would be exactly the same until you manually changed its data.
## **The self Parameter**
The first parameter of __init__ must always be self. It represents the specific instance of the object being created. It allows the method to attach data (attributes) to that particular object.

Key Rules
* Automatic Execution: You never call obj.__init__() manually; Python does it when you write ClassName().
* No Return Value: The __init__ method cannot return any value (like a string or number); its only job is to set up the object.
* Customization: You can add as many parameters as you want after self to make each object unique.




```python
class Dog:
  def __init__(self):
    self.name = 'Buddy'
  def bark(self):
    print(f'{self.name} is barking')

my_dog = Dog()
my_dog.bark()


class Car:
  def __init__(self,color,brand):
    self.color = color
    self.brand = brand
my_car = Car('Red','Toyota')
print('This for car 1', my_car.color)
print('This for car 1',my_car.brand)

my_car.color = 'White'
my_brother_car = Car('Blue','Honda')
print('This for brother car 1', my_brother_car.color)
print('This for brother car 1', my_brother_car.brand)
print('change color of car 1', my_car.color)
```

## **Instance Variables**
These are variables that belong to a specific object. Every object created from the same class can have different values for these variables.
* Defined in: Usually inside the __init__ constructor using self.variable_name.
* Scope: They are unique to each instance. Changing the value for one object does not affect others.

## **Instance Methods**
These are functions defined inside a class that can access and modify an object's instance variables.
* The self Requirement: They must take self as their first parameter so they know which object’s data to work with.
* Usage: They represent the actions or behaviors of a specific object.


```python
class student():
  def __init__(self,name,age,grade):
    self.name = name
    self.age = age
    self.grade = grade

  def desplay_info(self):
    print(f'Name: {self.name}, Age: {self.age}, Grade: {self.grade}')
    print(f'Student name: {self.name}')
    print(f'Student age: {self.age}')
    print(f'Student grade: {self.grade}')

  def Eligible(self):
    if self.grade >= 50:
      print(f'{self.name} is eligible')
    else:
      print(f'{self.name} is not eligible')


student_1 = student('Mr.X',20,60)
student_1.desplay_info()
student_1.Eligible()

student_2 = student('Mr.Y',20,40)
student_2.desplay_info()
student_2.Eligible()


# Accessing Attributes
print(student_1.name)
print(student_2.age)

# Calling Method
student_1.desplay_info()
student_2.Eligible()
```

## **Encapsulation in Python**
Encapsulation is the practice of bundling data and methods into a single unit (a class) and restricting access to the inner workings of that object. This prevents accidental data corruption and hides complexity.
## **Public Attributes**
By default, all attributes in a Python class are public. They can be accessed and modified from anywhere—both inside and outside the class.
* Naming: No underscores (e.g., self.name).
* Access: object.name works perfectly.
## **Private Attributes (Strong Internal Use)**
Private attributes are intended to be hidden from the outside world. Python "mangles" the name internally to make it harder (but not impossible) to access from outside.
* Naming: Two leading underscores (e.g., self.__salary).
* Access: Attempting object.__salary will result in an AttributeError.

## **Protected Attributes (A Gentle Warning)**
There is also a "Protected" level. It doesn't actually block access; it’s just a convention among programmers that says, "Please don't touch this outside of this class or its subclasses."
* Naming: One leading underscore (e.g., self._tax_id).

## **Use Encapsulation**'
* **Control:** You can use "Getter" and "Setter" methods to validate data before changing a private variable (e.g., ensuring a salary isn't negative).
* **Security:** It protects the internal state of the object from unintended interference.

```python
class student():
  def __init__(self,name,grade):
    self.__grade = grade

  def get_grade(self):
    return self.__grade

student = student('Mr.X',50)
print(student.get_grade())
```

Getters and setters are methods used to retrieve (get) and update (set) the value of private attributes. They act as "security guards," allowing you to add validation logic before data is accessed or changed.

# **Inheritance in Python**
Inheritance allows one class (the Child) to derive or "inherit" attributes and methods from another class (the Parent). It is a key tool for reusing code and building a hierarchy.
### **The Parent Class (Base Class)**
The parent class is the existing class with the general features you want to share.
Example: A Vehicle class that has a brand and a start() method.
### **The Child Class (Derived Class)**
The child class inherits everything from the parent but can also add its own specific features or modify existing ones.
Syntax: class ChildClass(ParentClass):


```python
class animals():
  def speak(self):
    print('some sound')

class dog(animals):
  pass

dog_1 = dog()
dog_1.speak()

class dog(animals):
  def bark(self):
    print('woof')

child_dog = dog()
child_dog.bark()
child_dog.speak()
```

## **Method Overriding**
Method Overriding occurs when a Child Class provides a specific implementation for a method that is already defined in its Parent Class.

It allows the child class to change or "refine" the behavior of a inherited method to better suit its needs.


```python
class animals():
  def speak(self):
    print('some sound')

class dog(animals):
  def speak(self):
    print('woof woof')

dog_1 = dog()
dog_1.speak()
```

## **Polymorphism in Python**
Polymorphism (meaning "many forms") allows different classes to be treated as instances of the same general class through the same interface. In Python, this is primarily achieved through overriding and overloading.
### **Method Overriding (Run-time Polymorphism)**
As we discussed, this is when a child class provides a specific implementation of a method already defined in its parent. It allows you to call the same method name on different objects and get different results.
### **Method Overloading (Compile-time Polymorphism)**
Method overloading is the ability to define multiple methods with the same name but different parameters (number or type).

**In Other Languages (like Java/C++):** You can have two distinct functions named add(int a, int b) and add(int a, int b, int c).

**In Python:** Python does not support traditional overloading. If you define two functions with the same name, the second one will simply overwrite the first.
### **Operator Overloading**
Python also allows you to change how operators (like +, -, *) behave with your custom objects using special "Magic Methods."

**Example:** Defining __add__ in a class allows you to use the + sign between two objects of that class.

## **Special methods**
Special methods (also known as Magic Methods or Dunder Methods) are predefined methods in Python that start and end with double underscores, like __init__.

They allow you to emulate built-in behavior or implement operator overloading for your custom objects. You don't usually call them directly; Python calls them "behind the scenes" when you perform specific actions (like adding two objects or printing one).
### **The Basics: Initialization and Representation**
These control how an object is created and how it looks when printed.
* __init__(self, ...): The constructor. Initializes a new object.
* __str__(self): Controls what happens when you print(obj) or use str(obj). It should return a user-friendly string.
* __repr__(self): Returns an "official" string representation of the object, mainly used for debugging.
### **Operator Overloading Methods**
These allow you to use symbols like +, -, or * with your custom objects.

* Operator (+)	 __add__(self, other)	(obj1 + obj2)
* Operator (-)	 __sub__(self, other)	(obj1 - obj2)
* Operator (*)	 __mul__(self, other)	(obj1 * obj2)
* Operator (==)  __eq__(self, other)	(obj1 == obj2)
* Operator <()	 __lt__(self, other)	(obj1 < obj2)
### **Length and Collection Methods**
If you want your object to behave like a list or a string.
* __len__(self): Allows you to use len(obj).
* __getitem__(self, key): Allows you to use square brackets like obj[index].

```python
class book():
  def __init__(self,title,auther):
    self.title = title
    self.auther = auther
  def __str__(self):
    return f'{self.title} by {self.auther}'

b = book('1984','George Orwell')
print(b)



class playlist():
  def __init__(self,songs):
    self.songs = songs
  def __len__(self):
    return len(self.songs)
  def __add__(self,other):
    return playlist(self.songs + other.songs)

p_1 = playlist(['song1','song2'])
p_2 = playlist(['song3'])
print(len(p_1))
print(len(p_1 + p_2))


class point():
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def __str__(self):
    return f'({self.x},{self.y})'
  def __add__(self,other):
    if not isinstance(other,point):
      return NotImplemented
    return point(self.x + other.x,self.y + other.y)

point_1 = point(1,2)
point_2 = point(3,4)

print(point_1 + point_2)
```

## **Reading from & Writing to Files in Python**

File handling in Python involves opening a file, performing operations like reading or writing, and closing it to free up resources.

### **1. Opening and Closing Files**
The standard and safest method is the with open() statement. It acts as a context manager, automatically closing the file for you even if an error occurs.

**Syntax:** with open("filename.txt", "mode") as file:

**Manual Close:** If you use file = open(), you must explicitly call file.close().

### **2. Common File Modes**
The mode specifies how you intend to interact with the file.

* 'r' (Read): Default mode. Opens for reading; error if the file doesn't exist.
* 'w' (Write): Opens for writing. It overwrites existing content or creates a new file.
* 'a' (Append): Adds data to the end of an existing file without deleting its content.
* 'x' (Create): Creates a new file; returns an error if the file already exists.
* 'b' (Binary): Used for non-text files like images (e.g., 'rb' or 'wb').

### **3. Reading Methods**
Python provides three primary ways to extract data:

* read(): Returns the entire file as a single string.
* readline(): Reads exactly one line at a time.
* readlines(): Reads all lines and returns them as a list of strings.
* Iteration: For large files, it is most memory-efficient to loop directly over the file object: for line in file:.

### **4. Writing Methods**
* **write("string"):** Inserts a specific string into the file. You must manually add \n if you want a new line.

* **writelines(list):** Writes a list of strings to the file all at once.

## **Modes of File Handling - Read, Write & Append**
In Python, the mode you choose when opening a file determines what you are allowed to do with it and what happens to the existing data.
### **Read Mode ('r')**
This is the default mode. Use it when you only need to pull information out of a file.

### **Append Mode ('a')**
Use this when you want to add new data to the end of a file without losing what is already there.
* Behavior:
1. If the file exists: The file pointer is placed at the end of the file. New data is added after the existing content.
2. If the file doesn't exist: It creates a new file.
* Code: with open("log.txt", "a") as f:

## **Using Context Managers (with Statement)**
A Context Manager is a Python construct that allows you to manage resources efficiently, ensuring they are properly set up and cleaned up. The most common way to use a context manager is with the with statement.
### **Why use the with statement?**
When you work with resources like files, network connections, or database locks, you must close them when you’re done. If your code crashes before reaching file.close(), the file stays "locked" or uses up memory.

The with statement guarantees that the resource is closed automatically, even if an error occurs inside the block.
### **Basic Example: File Handling**
Compare the "Old Way" vs. the "Modern Way"

### **How it Works (Under the Hood)**
A context manager relies on two "Magic Methods" within a class:
* __enter__: Sets up the resource (e.g., opens the file) and assigns it to the variable after as.
* __exit__: Cleans up the resource (e.g., closes the file) when the code block finishes or fails.
### **Common Use Cases**
* File I/O: Ensuring files aren't corrupted or left open.
* Database Connections: Opening a connection and ensuring it disconnects properly.
* Locks in Multithreading: Acquiring a lock and releasing it so other threads aren't blocked forever.
* Temporary Contexts: Changing a setting (like decimal precision) and having it revert back automatically.
### **Summary of Benefits**
* Cleaner Code: You don't need try...finally blocks everywhere.
* Safety: Prevents memory leaks and file corruption.
* Readability: Clearly defines the "scope" where a resource is being used.

```python
with open('/demo.txt', 'r') as file:
  data = file.read()
  print(data)

# file close here automatically
```

* **IOError:** A general error for any Input/Output failure (like a disconnected drive).

## **Introduction to Exceptions in Python**
An exception is an error that happens while a program is running (at "runtime"). Even if your code is written with perfect syntax, it can still "crash" if it encounters an unexpected situation, like trying to divide by zero or opening a file that isn't there. [2]
### **Exception vs. Syntax Error**
Syntax Error: These are "grammar" mistakes (like a missing colon :) that stop the program before it even starts. [1]
Exception: The code is technically correct, but an error occurs during execution (e.g., 10 / 0). [2]
### **Common Built-in Exceptions**
Python has many built-in types to help you identify what went wrong: [4]
* **ZeroDivisionError:** Trying to divide a number by zero.
* **TypeError:** Performing an operation on the wrong data type (like adding a string to an integer).
* **ValueError:** A function receives an argument of the right type but an inappropriate value (e.g., int("abc")).
* **IndexError:** Trying to access an index that is out of range in a list.
* **KeyError:** Trying to access a dictionary key that doesn't exist.
### **Why Handle Exceptions?**

If an exception isn't "caught," Python stops the program and displays a traceback (an error message). By handling exceptions, you can: [3]
* Keep the program running instead of crashing.
* Provide user-friendly error messages.
* Clean up resources (like closing files) before the program stops.
### **Basic Structure**
Exceptions are managed using the try and except keywords.

## **Using try, except, else & finally Blocks**
In Python, these four blocks create a complete system for handling errors and ensuring your code runs reliably.
### **The try Block**
This is where you place the code that might cause an error. Python "tries" to run this first.
### **The except Block**
This block executes only if an error occurs inside the try block. You can catch specific errors (like ValueError) or any error using a general Exception.
### **The else Block**
This block runs only if no errors occurred in the try block. It is useful for code that should only run if the "try" was successful.
### **The finally Block**
This block runs no matter what, whether an error happened or not. It is primarily used for cleanup tasks, like closing a database connection or a file.

```python
num_1 = int(input())
num_2 = int(input())
print(num_1/num_2)
```

```python
try:
  num_1 = int(input())
  num_2 = int(input())
  print(num_1/num_2)
except ZeroDivisionError:
  print('you can not divide by zero')
except ValueError:
  print('invalid input')
except Exception as e:
  print(f'An unexpected error occured: {e}')
```

```python
try:
  print('Running code....')
except:
  print('Error found')
else:
  print('No error found')
finally:
  print('Done')
```

## **Raising Exceptions in Python**
In Python, you don't have to wait for a built-in error to happen; you can intentionally trigger an exception yourself using the raise keyword. This is useful when a program is technically "correct" but breaks your specific business rules (e.g., a bank withdrawal larger than the balance).

**Basic Syntax:**
You simply use the raise keyword followed by the name of the exception and an optional custom message.

```python
age = -3
if age < 0:
  raise ValueError('Age cannot be negative')
```

## **Creating Custom Exceptions in Python**
To create a custom exception, you define a new class that inherits from Python's built-in Exception class. This allows you to create specific error types tailored to your application's needs.

You typically only need a few lines of code. The pass keyword is used because you are just naming the error; the parent Exception class handles the heavy lifting.

**Why Create Custom Exceptions?**
* Clarity: InvalidPasswordError is much more descriptive than a generic ValueError.
* Targeted Catching: You can catch only your specific business errors without accidentally catching other built-in errors.
* API Design: If you are building a library for others, custom exceptions help them understand exactly what went wrong in your specific logic.

```python
class InvalidPasswordError(Exception):
  pass

age = int(input('Enter yout age: ' ))
if age < 0:
  raise InvalidPasswordError(f'Age cannot be negative: {age}')
```

```python
try:
  age = int(input('Enter yout age: ' ))
  if age < 0:
    raise InvalidPasswordError(f'Age cannot be negative: {age}')
except InvalidPasswordError as e:
  print(e)
```
